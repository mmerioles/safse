```python
from tensorflow.keras.models import load_model

from fish_classifier.data import prepare_data
from fish_classifier.predict import predict_labels
from fish_classifier.model import build_model
```

```python
(
    X_train,
    y_train,
    X_val,
    y_val,
    X_test,
    y_test_enc,
    y_test,
    le, 
) = prepare_data("../data/folds-w-meta.pkl")

model = build_model(len(le.classes_))

model.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=32,
)
```

```python
model.save("models/classifier.keras")
```

```python
import numpy as np

test_loss, test_acc = model.evaluate(X_test, y_test_enc)

print("Test Accuracy:", test_acc)

pred_probs = model.predict(X_test)
pred_classes = np.argmax(pred_probs, axis=1)
pred_labels = le.inverse_transform(pred_classes)

print(pred_labels[:10])
print(y_test[:10])
```
